```python
#Function to get outliers in a specific column using the IQR method
def get_ouliers(df , column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    IQR = Q3 - Q1

    lower_fence = Q1 - 1.5 * IQR
    upper_fence = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_fence) | (df[column] > upper_fence)]
    return outliers
```

```python
income_outliers = get_ouliers(df , 'MonthlyIncome')
len(income_outliers)
```


##  Summary & Insights

### Dataset Overview
The IBM HR Analytics dataset contains **1,470 employee records** with **35 original features**. After data cleaning, **4 redundant columns** were removed — `EmployeeNumber`, `EmployeeCount`, `Over18`, and `StandardHours` — as they either held constant values or offered no predictive value.

---

### Data Cleaning Summary
- **No missing values** were found in the dataset, making imputation unnecessary.
- **Outliers** were detected in `MonthlyIncome`, `YearsAtCompany`, `TotalWorkingYears`, and `NumCompaniesWorked` using the IQR method. All were retained as they represent valid real-world variation — senior professionals, long-tenured employees, and high-income roles.
- The final cleaned dataset is **structurally sound and ready for next step**.

---

### Key EDA Insights

**1. Age Distribution**
The workforce is predominantly between **25–40 years old**, forming a near-normal distribution with a slight right skew. This indicates a relatively young-to-mid-career workforce.

**2. Monthly Income**
Income is heavily **right-skewed** — most employees earn in lower bands while a small group earns significantly more. Employees who left tend to cluster in lower income brackets, pointing toward compensation as a contributing attrition factor.

**3. Years at Company**
Most employees have **0–7 years of tenure**. Attrition is highest in the early years, suggesting onboarding gaps, misaligned expectations, or better external opportunities pulling employees away early in their journey.

**4. Department-wise Attrition**
The **R&D department** has the highest absolute attrition, which is partially explained by it also being the largest department. Relative attrition rates should be examined further in modeling.

**5. Overtime & Attrition**
Employees who work **overtime show noticeably higher attrition**. This is a strong signal — overwork correlates with burnout and dissatisfaction, making it one of the more actionable levers for HR intervention.

**6. Mean Comparison (Attrition vs. Non-Attrition)**

| Feature | Attrition (Yes) | Attrition (No) |
|---|---|---|
| Age | Lower | Higher |
| Monthly Income | Lower | Higher |
| Years at Company | Fewer | More |
| Job Satisfaction | Slightly lower | Slightly higher |

**7. Correlation Highlights**
- `MonthlyIncome` & `JobLevel` → **Strong positive correlation** — senior roles command higher pay.
- `TotalWorkingYears` & `JobLevel` / `MonthlyIncome` → **High correlation** — experience drives both.
- `Age` & `TotalWorkingYears` → **Moderate correlation** — as expected.
- `JobSatisfaction` shows **weak correlations** with structural variables, suggesting it is independently influenced.

---

## Conclusion

This EDA reveals that employee attrition is **not driven by a single factor** but by a combination of structural and experiential variables. The most important takeaways are:

1. **Early tenure is the highest-risk window.** The majority of attrition happens within the first 5 years. Targeted onboarding, mentorship, and career development programs during this period could significantly improve retention.

2. **Compensation matters, but isn't the whole story.** While lower-income employees leave more often, boxplots show considerable overlap between groups — meaning pay alone doesn't fully explain attrition. Role fit, growth opportunity, and work environment likely play equally important roles.

3. **Overtime is a red flag.** The strong association between overtime and attrition suggests that workload management and work-life balance policies deserve direct attention from HR leadership.

4. **Job satisfaction gap is modest.** The difference in satisfaction scores between leavers and stayers is small, implying employees may leave even when somewhat satisfied — driven more by opportunity elsewhere or structural dissatisfaction.

5. **Feature correlations are interpretable and clean.** The strong links between `JobLevel`, `MonthlyIncome`, and `TotalWorkingYears` are logically sound and will be useful features in any predictive model going forward.
